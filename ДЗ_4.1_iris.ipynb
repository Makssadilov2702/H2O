```python
import h2o
h2o.init()

# -1-
datasets = "https://raw.githubusercontent.com/DarrenCook/h2o/bk/datasets/"
data = h2o.import_file(datasets + "iris_wheader.csv")
y = "class"
x = data.names
x.remove(y)

# определяем параметры обучения, входные и целевые параметры
train, test = data.split_frame([0.8])

# cтроит многослойную искусственную нейронную сеть с прямой связью на H2OFrame
m = h2o.estimators.deeplearning.H2ODeepLearningEstimator()

# обучаем сеть
m.train(x, y, train)

# выполним предсказания на основе модели
p = m.predict(test)
```

```python
m.train(x, y, train)
```

```python
p.as_data_frame()
```

```python
#получим долю правильных ответов
(p["predict"]== test["class"]).mean()
```

```python
#вывод результата в виде 2-х столбцов с предсказаниями и  правильными метками классов
p["predict"].cbind(test["class"]).as_data_frame()
```

```python
from h2o.estimators import H2ORandomForestEstimator
```

```python
# -2-
#  определяем модель распределенного случайного леса (DRF) для регрессии/ классификации
model = H2ORandomForestEstimator(ntrees=50, max_depth=20, nfolds=10)

# обучаем модель классификации
model.train(x, y, training_frame=train)
```

```python
# проверка производительности модели
performance = m.model_performance(test_data=test)
performance
```

```python
# выполним предсказания на основе модели
p1 = model.predict(test)
```

```python
p1["predict"].cbind(test["class"]).as_data_frame()
```
